## Pre-processing
> It is required to convert features that are extracted from `pcap` as hexadecimal

> Replaced the fields with null values (NaN) by zero

```python
fields = ['eth.type', 'ip.id', 'ip.flags', 'ip.checksum', 'ip.dsfield', 'tcp.flags', 'tcp.checksum']

for field in fields:
    df[field] = df[field].apply(lambda x: int(str(x), 16))
    
```

```python
bonafide = bonafide.fillna(0)
for field in fields:
    bonafide[field] = bonafide[field].apply(lambda x: int(str(x), 16))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier

algorithms = {
    "MLP" : (MLPClassifier(random_state=17), {
        "hidden_layer_sizes" : (10, 10),
    }),
    "SVM" : (LinearSVC(random_state=17), {}),
    "KNN" : (KNeighborsClassifier(n_jobs=-1), {
        "n_neighbors" : [1, 3, 5]
    }),
    "XGB" : (XGBClassifier(random_state=17, n_jobs=-1), {}),
    "NB" : (GaussianNB(), {}),
    "LR" : (LogisticRegression(random_state=17, n_jobs=-1), {}),
    "RF" : (RandomForestClassifier(random_state=17, n_jobs=-1), {
        "n_estimators" : [10, 15, 20],
        "criterion" : ("gini", "entropy"), 
        "max_depth": [5, 10],
        "class_weight": (None, "balanced", "balanced_subsample")
    }),
    "DT" : (DecisionTreeClassifier(random_state=17), {
        "criterion": ("gini", "entropy"), 
        "max_depth": [5, 10, 15],
        "class_weight": (None, "balanced")
    }),
}

full_data = full_data.fillna(0)
X = full_data.drop(columns = ["label"])
y = full_data.label

print(X.shape, y.shape)
```

```python
X.head()
```

```python
X = X.astype(int) 
X.head()
```

## Grid Search for Machine Learning Algorithms

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, make_scorer, f1_score, roc_auc_score

kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=17) # Train, Test
gskf = StratifiedKFold(n_splits=3, shuffle=True, random_state=17) # Validation
perf = f1_score # could be considered roc_auc_score for binary classification

score = {}
best_params = {}
for algorithm in algorithms.keys():
    score[algorithm] = []

for algorithm, (clf, parameters) in algorithms.items():
    print(algorithm)
    for train, test in kf.split(X, y):
        prep = StandardScaler()
        #prep = MinMaxScaler()
        prep.fit(X.iloc[train])
        best = GridSearchCV(clf, parameters, cv=gskf, scoring=make_scorer(perf), n_jobs=-1)
        best.fit(prep.transform(X.iloc[train]), y.iloc[train])
        best_params[algorithm] = best.best_params_
        score[algorithm].append(perf(best.predict(prep.transform(X.iloc[test])), y.iloc[test]))
```

```python
# f1-scores for the best set of parameters from the Grid Search above (for each k-fold)
pd.DataFrame.from_dict(score)
```

# ROC/AUC Evaluation

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, make_scorer, f1_score, roc_auc_score, roc_curve

kf = StratifiedKFold(n_splits=10, shuffle=True, random_state=17) # Train, Test
gskf = StratifiedKFold(n_splits=3, shuffle=True, random_state=17) # Validation
perf = roc_auc_score

score = {}
best_params = {}
roc = {}

for algorithm in algorithms.keys():
    score[algorithm] = []
    roc[algorithm] = {}

for algorithm, (clf, parameters) in algorithms.items():
    print(algorithm)
    for train, test in kf.split(X, y):
        prep = StandardScaler()
        #prep = MinMaxScaler()
        prep.fit(X.iloc[train])
        best = GridSearchCV(clf, parameters, cv=gskf, scoring=make_scorer(perf), n_jobs=-1)
        best.fit(prep.transform(X.iloc[train]), y.iloc[train])
        best_params[algorithm] = best.best_params_
        score[algorithm].append(perf(best.predict(prep.transform(X.iloc[test])), y.iloc[test]))
        fpr, tpr, thr = roc_curve(y.iloc[test], best.predict_proba(prep.transform(X.iloc[test])).transpose()[1])
        roc[algorithm]['fpr'] = fpr
        roc[algorithm]['tpr'] = tpr
        roc[algorithm]['threshold'] = thr
```

```python
# ROC/AUC scores for the best set of parameters from the Grid Search above (for each k-fold)
pd.DataFrame.from_dict(score)
```

```python
plt.style.use('plot_style.txt')
```

```python
import os
plt.figure()

index = 0
for model_key, _ in roc.items():
    AUC = pd.DataFrame.from_dict(score).mean().round(decimals=3)[index]
    plt.plot(roc[model_key]['fpr'], roc[model_key]['tpr'], label="{} (AUC={})".format(model_key, AUC))
    index = index + 1
    # save ROC data
    filename = 'data/Internet_'+model_key+'_ROC_data.csv'
    pd.DataFrame.from_dict(data=roc[model_key]).to_csv(filename, index=False)
    
plt.plot([0,1], [0,1], color='black', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC)')
plt.legend(loc="lower right")

plt.savefig(os.path.join('images/', 'roc_internet_models.pdf'), dpi=300, bbox_inches="tight")
```
